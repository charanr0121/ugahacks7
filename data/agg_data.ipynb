```python
import glob 
import pandas as pd 
```

```python
csv_files = glob.glob("history/*.csv")
files = [pd.read_csv(csv) for csv in csv_files]

df = pd.concat(files, axis = 0)
df.to_csv("master.csv")
```

```python
csv_files = glob.glob("crypto/Top 100 Crypto Coins/*.csv")
data_frames = []
for file in csv_files: 
    name = file[file.find("crypto/Top 100 Crypto Coins/") + len("crypto/Top 100 Crypto Coins/") + 1:file.find(".csv")]
    name = ((name.replace(" ", "_")))
            
    df = pd.read_csv(file)
    df["name"] = name
            
    data_frames.append(df)

df = pd.concat(data_frames, axis = 0)
df
```

```python
df[["Date", "Open", "High", "Low", "Close", "Volume", "Currency", "name"]].to_csv("all_crypto.csv", index = False)
```
